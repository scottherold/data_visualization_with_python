```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
data = pd.read_csv('fandango_score_comparison.csv')
```

```python
data.head()
```

```python
# Basic Scatter Plot
data.plot.scatter(x='RT_norm', y='RT_user_norm')
plt.plot((0, 5),(0, 5), 'k--')
```

```python
# Multiple Scatter Plots
data.plot.scatter(x=['Metacritic_norm', 'RT_norm'], y=['Metacritic_user_nom', 'RT_user_norm'], color=['b', 'r'])
plt.plot((0, 5), (0, 5), 'k--')
```

```python
# Seperate plot calls for more control
data.plot.scatter(x='RT_norm', y='RT_user_norm', color='r', marker='x')
# ax allows you to layer two plots together; plt.gca() overlays the first plot onto the plot where the object/method is called
data.plot.scatter(x='Metacritic_norm', y='Metacritic_user_nom', color='g', s=50, ax=plt.gca())
plt.xlabel('Critic Score')
plt.ylabel('USer Score')
plt.plot((0, 5), (0, 5), 'k--')
```
