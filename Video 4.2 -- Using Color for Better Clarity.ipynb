```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
data = pd.read_csv('fandango_score_comparison.csv')
```

```python
# Choosing an appropriate colormap
# A list of colormaps can be found in the matplotlib documentation
data.plot.scatter(y='RottenTomatoes', x='IMDB_user_vote_count', c="RottenTomatoes_User", vmin=0, vmax=100,
                 edgecolor='None', cmap='inferno')
plt.axhline(50, color='k', linestyle='--')
plt.ylim(0, 100)
```

```python
# Selecting colors from a colormap
cm = plt.cm.get_cmap('inferno')
for i in range(20):
    # the colormap uses an index, which is why you would use color=cm(i / 20)
    # the 20 in cm(i / 20) is used due to the loop using range(20)
    plt.plot(np.arange(0, 10, 0.1), np.sin(i * np.pi / 20 + np.arange(0, 10, 0.1)), color=cm(i / 20))
```
