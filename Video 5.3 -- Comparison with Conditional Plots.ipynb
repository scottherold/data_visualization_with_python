```python
import numpy as np
import pandas as pd
import seaborn as sns
import mtplotlib.pyplot as plt
%matplotlib inline
```

```python
# load data
booze = pd.read_csv('drinks.csv')
```

```python
booze.head()
```

```python
# violin plot
drinks = booze[['beer_servings', 'spirit_servings', 'wine_servings']]
sns.violinplot(data=drinks)
```

```python
# Limit data range
sns.violinplot(data=drinks, cut=0)
```

```python
# Swarm plot
sns.swarmplot(data=drinks)
```

```python
# Combined!
sns.violinplot(data=drinks, cut=0)
# can change the color and opacity of any seaborn graph
sns.swarmplot(data=drinks, color='grey', alpha=0.5)
```

```python
# Compare all relations with pairplot
sns.pairplot(booze)
```

```python
# Change plot type
sns.pairplot(booze, diag_kind='kde', kind='reg')
```
