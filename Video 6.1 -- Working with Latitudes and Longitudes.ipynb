```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
%matplotlib inline
# Set up figure size and DPI for screen demo
plt.rcParams['figure.figsize'] = (6, 4)
plt.rcParams['figure.dpi'] = 150
```

```python
# function to plot map
def setup_map(m):
    m.drawcoastlines()
    m.fillcontinents(color='olive', lake_color='aqua')
    # dara parallels and meridians
    m.drawmapboundary(fill_color='aqua')
    for y in np.linspace(m.ymax / 22, 19 * m.ymax / 22, 9):
        for x in np.linspace(m.xmax / 22, 19 * m.xmax / 22, 12):
            lon, lat = m(x, y, inverse=True)
            m.tissot(lon + 5, lat + 5, 1.5, 100, facecolor='k', zorder=10, alpha=0.5)
```

```python
# creating the most basic projection
m = Basemap(llcrnrlat=-80, urcrnrlat=80, llcrnrlon=-180, urcrnrlon=180, lat_ts=0)
setup_map(m)
# this simple map does not preserve SHAPE or AREA
```

```python
# Mercator projection
m = Basemap(projection='merc', llcrnrlat=-80, urcrnrlat=80, llcrnrlon=-180, urcrnrlon=180, lat_ts=0)
setup_map(m)
# Preserves SHAPE but not AREA
```

```python
# Equal Area projetion (Hammer projection)
m = Basemap(projection='hammer', lon_0=0)
setup_map(m)
# Preserves AREA but not SHAPE
```

```python
# Perspective projection
# satellite_height can use expotential notation, 4e5 == 4 to the power of 5
m = Basemap(projection='nsper', lon_0=-114, lat_0=51, satellite_height=4e5, resolution='l')
m.drawcoastlines()
m.drawmapboundary(fill_color='aqua')
m.fillcontinents(color='olive', lake_color='aqua')
```
