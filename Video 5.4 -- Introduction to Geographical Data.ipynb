```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
%matplotlib inline
```

```python
# Generate a Mercator Projection
# Doesn't generate any plots, but generates an object which you can draw plots from
m = Basemap(projection='merc', llcrnrlat=-80, urcrnrlat=80, llcrnrlon=-180, urcrnrlon=180, lat_ts=20)
```

```python
# Draw coastlines
m.drawcoastlines()
```

```python
# Draw countries
m.drawcoastlines()
m.drawcountries()
```

```python
# Fill water/land
m.fillcontinents(color='olive', lake_color='blue')
m.drawmapboundary(fill_color='aqua')
m.drawrivers(color='navy')
```

```python
# Relief map
m = Basemap(projection='merc', llcrnrlat=30, urcrnrlat=50, llcrnrlon=-15, urcrnrlon=15, lat_ts=40)
m.shadedrelief()
```

```python
# 3D Orthographic projection
m = Basemap(projection='ortho', lon_0=8, lat_0=30)
m.fillcontinents(color='olive', lake_color='aqua')
m.drawmapboundary(fill_color='aqua')
```

```python
# 3D Orthographic projection with pretty satellite image
m = Basemap(projection='ortho', lon_0=8, lat_0=30)
m.bluemarble()
```
