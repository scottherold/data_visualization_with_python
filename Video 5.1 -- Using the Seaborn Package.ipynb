```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
data = pd.read_csv('fandango_score_comparison.csv')
```

```python
data.head()
```

```python
# Distribution plot
# Kernal Density is the curved line (KDE)
sns.distplot(data['RT_norm'])
```

```python
# KDE plot
sns.kdeplot(data['RT_norm'], shade=True)
```

```python
# 2D KDE plot
sns.kdeplot(data['RT_norm'], data['RT_user_norm'], shade=True)
```

```python
# The Joint plot
# automatically provides a correlation coefficient == pearson's r (probably that the coefficient is random)
sns.jointplot(x='RT_norm', y='RT_user_norm', data=data, kind='hex')
```

```python
# Theming Seaborn
# set_style() alters the graph's appearance
sns.set_style('ticks')
sns.jointplot(x='RT_norm', y='RT_user_norm', data=data, kind='kde', shade=False)
```

```python
# Spine Control
sns.set_style('whitegrid')
sns.jointplot(x='RT_norm', y='RT_user_norm', data=data, kind='kde', shade=False)
# remove spines
sns.despine(left=True, bottom=True)
```

```python
# Automatic Scaling
sns.set_style('whitegrid')
sns.set_context('notebook') # automatic scaling
sns.jointplot(x='RT_norm', y='RT_user_norm', data=data, kind='kde', shade=False)
```
