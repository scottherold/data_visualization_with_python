```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# Generate some mock data
from scipy.ndimage.filters import gaussian_filter
x = np.random.normal(size=1000)
y = np.random.normal(size=1000)
z = np.random.normal(size=1000)
im = gaussian_filter(np.random.normal(size = (512, 512)), 16)
```

```python
# Show available stylesheets
plt.style.available
```

```python
# Stylesheets
# grayscale only -- good for printing on paper black and white
# plt.style.use('grayscale')

# gg plot (from R)
# plt.style.use('ggplot')

plt.subplot(221)
plt.plot(np.arange(0, 20, 0.1), np.sin(np.arange(0, 20, 0.1)), 'o-')
plt.subplot(222)
plt.scatter(x, y, c=z)
plt.subplot(223)
plt.hist((x, y, z))
plt.subplot(224)
plt.imshow(im)
plt.gcf().set_size_inches(12, 8)
```

```python
# LaTeX Labels
plt.hist((-x**2, -y**2, -z**2))
plt.xlabel('x')
plt.ylabel(r'$x^2$') # Mathematical language
```

```python
# Font properties
plt.rcParams['font.size'] = 12
plt.hist((-x**2, -y**2, -z**2))
plt.xlabel('x')
plt.ylabel('x^2')
```

```python
# Tweaking the legend
# legend defaults to best
plt.hist((-x**2, -y**2, -z**2), label=('x', 'y', 'z'))
# plt.legend(loc='upper left')

# can modify the legend to make background transparent
plt.legend(loc='upper right', framealpha=0.5)
plt.xlabel('x')
plt.ylabel('x^2')
```
