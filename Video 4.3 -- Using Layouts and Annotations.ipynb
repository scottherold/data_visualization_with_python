```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
data = pd.read_csv('fandango_score_comparison.csv')
```

```python
# Add Vertical and Horizontal Marker Lines
data.plot.scatter(x='Metacritic', y='Metacritic_User')
# creates horizontal line at 5
plt.axhline(5, color='k')
# creates vertical line at 50
plt.axvline(50, color='k')
```

```python
# add Text
data.plot.scatter(x='Metacritic', y='Metacritic_User')
# label for movies fans like, but critics do not; numbers are positioning x/y
plt.text(10, 9, 'Popcorn Pleasers')
# label for movies that critics like, but fans do not; numbers are positioning x/y
plt.text(70, 4, 'Snob Favorites')
# creates horizontal line at 5
plt.axhline(5, color='k')
# creates vertical line at 50
plt.axvline(50, color='k')
```

```python
# find location at point
data.loc[data['Metacritic'] == data['Metacritic'].min()]
```

```python
# user the annotate() method
data.plot.scatter(x='Metacritic', y='Metacritic_User')
# label for movies fans like, but critics do not; numbers are positioning x/y
plt.text(10, 9, 'Popcorn Pleasers')
# label for movies that critics like, but fans do not; numbers are positioning x/y
plt.text(70, 4, 'Snob Favorites')
# creates horizontal line at 5
plt.axhline(5, color='k')
# creates vertical line at 50
plt.axvline(50, color='k')
# annotates a specific point; xy == point of interest, xytext == text location
# arrowprops creates an arrow from the xytext to the xy position
plt.annotate('Paul Blart 2', xy=(13, 2.4), xytext=(10, 4), arrowprops=dict(facecolor='r', shrink=0.2))
```
