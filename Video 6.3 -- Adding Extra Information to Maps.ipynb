```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
%matplotlib inline
# Set up figure size and DPI for screen demo
plt.rcParams['figure.figsize'] = (6, 4)
plt.rcParams['figure.dpi'] = 150
```

```python
# function to render map
def setup_map(m):
    m.drawcoastlines(zorder=-1)
    m.fillcontinents(color='olive', lake_color='aqua', zorder=-1)
    # draw parallels and meridians
    m.drawmapboundary(fill_color='aqua')
    plt.gcf().set_size_inches(12, 10)
```

```python
# Add Parellels & Meridians
m = Basemap(width=1.2e7, height=9e6, projection='lcc', lat_1=45, lat_2=55, lat_0=49.5, lon_0=8.7)
# labels put position as if on a graph
m.drawmeridians(np.arange(0,360,30), labels=[1, 0, 0, 1]) # from 0 to 360 degrees in steps of 30
# labels put position as if on a graph
m.drawparallels(np.arange(0,90,30), labels=[0, 1, 1, 0])# from 0 to 90 degress in stesp of 30
setup_map(m)
```

```python
# Add Terminator (allows for date/time shading)
import datetime
m = Basemap(width=1.2e7, height=9e6, projection='lcc', lat_1=45, lat_2=55, lat_0=49.5, lon_0=8.7)
m.nightshade(datetime.datetime.now()) # shows night/date for current time
setup_map(m)
```

```python
# Add scale
m = Basemap(width=1.2e7, height=9e6, projection='lcc', lat_1=45, lat_2=55, lat_0=49.5, lon_0=8.7)
# adds scale label at the position provided, length in kilometers
m.drawmapscale(-20, 20, -20, 20, length=1000, barstyle='fancy', labelstyle='fancy')
setup_map(m)
```

```python
# Level of detail with resolution
m = Basemap(width=5e6, height=4e6, projection='lcc', lat_1=45, lat_2=55, lat_0=49.5, lon_0=8.7, resolution='i')
m.drawcountries()
setup_map(m)
```
