```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# load data
data = pd.read_csv('fandango_score_comparison.csv')
```

```python
# load check / column headers check -- first 5 rows
data.head()
```

```python
# Columns for comparison
scores = ['RT_norm', 'IMDB_norm', 'RT_user_norm']
```

```python
# Bar Chart w/ Pandas
data.head().plot.bar(x='FILM', y=scores)
plt.ylabel('score')
```

```python
# Horizontal Bars
data.head().plot.barh(x='FILM', y=scores)
plt.xlabel('score')
```

```python
# Bar Colors
data.head().plot.bar(x='FILM', y=scores, color=['y', 'r', 'k'])
plt.ylabel('score')
```

```python
# Bar Hatching
data.head().plot.bar(x='FILM', y=scores, hatch='x')
plt.ylabel('score')
```

```python
# Bar Width
data.head().plot.bar(x='FILM', y=scores, width=1)
plt.ylabel('score')
```

```python
# Stacked charts and logrithmic axis
data.head().plot.bar(x='FILM', y=['Metacritic_user_vote_count', 'Fandango_votes', 'IMDB_user_vote_count'],
                    stacked=True, log=True)
plt.ylabel('score')
```
